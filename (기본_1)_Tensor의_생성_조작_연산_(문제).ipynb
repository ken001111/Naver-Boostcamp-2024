````python
# ë§¨í•´íŠ¼ ìœ ì‚¬ë„ë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì •ì˜
def get_manhattan_similarity(t1, t2):
  # TODO 5-1) t1, t2 Tensorì˜ ë§¨í•´íŠ¼ ê±°ë¦¬ë¥¼ êµ¬í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•´ ì£¼ì„¸ìš”. | ë³€ìˆ˜ëª…: manhattan_distance
  # TODO 5-1
  manhattan_distance = torch.norm(t1 -t2, p = 1)

  # TODO 5-2) ë§¨í•´íŠ¼ ê±°ë¦¬ë¥¼ ì´ìš©í•˜ì—¬ ë§¨í•´íŠ¼ ìœ ì‚¬ë„ë¥¼ êµ¬í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•´ ì£¼ì„¸ìš”. | ë³€ìˆ˜ëª…: manhattan_similarity
  # TODO 5-2
  manhattan_similarity = 1 / (1 + manhattan_distance)
  print("manhattan_similarity: {}".format(manhattan_similarity))

  # TODO 5-3) ë§¨í•´íŠ¼ ìœ ì‚¬ë„ë¥¼ ë°˜í™˜í•´ ì£¼ì„¸ìš”.
  return manhattan_similarity# TODO 5-3
````

````python
# ìœ í´ë¦¬ë“œ ìœ ì‚¬ë„ë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì •ì˜
def get_euclidean_similarity(t1, t2):
  # TODO 5-4) t1, t2 Tensorì˜ ìœ í´ë¦¬ë“œ ê±°ë¦¬ë¥¼ êµ¬í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•´ ì£¼ì„¸ìš”. | ë³€ìˆ˜ëª…: euclidean_distance
  # TODO 5-4
  euclidean_distance = torch.norm(t1 - t2, p = 2)

  # TODO 5-5) ìœ í´ë¦¬ë“œ ê±°ë¦¬ë¥¼ ì´ìš©í•˜ì—¬ ìœ í´ë¦¬ë“œ ìœ ì‚¬ë„ë¥¼ êµ¬í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•´ ì£¼ì„¸ìš”. | ë³€ìˆ˜ëª…: euclidean_similarity
  # TODO 5-5
  euclidean_similarity = 1 / (1 + euclidean_distance)
  print("euclidean_similarity: {}".format(euclidean_similarity))

  # TODO 5-6) ìœ í´ë¦¬ë“œ ìœ ì‚¬ë„ë¥¼ ë°˜í™˜í•´ ì£¼ì„¸ìš”.
  return euclidean_similarity # TODO 5-6
````

````python
# ì½”ì‚¬ì¸ ìœ ì‚¬ë„ë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì •ì˜
def get_cosine_similarity(t1, t2):
  # TODO 5-7) t1, t2 Tensorì˜ ë‚´ì ì„ ì´ìš©í•˜ì—¬ ì½”ì‚¬ì¸ ìœ ì‚¬ë„ë¥¼ êµ¬í•˜ëŠ” ì½”ë“œë¥¼ ì‘ì„±í•´ ì£¼ì„¸ìš”. | ë³€ìˆ˜ëª…: cosine_similarity
  # TODO 5-7
  cosine_similarity = torch.dot(t1, t2) / (torch.norm(t1 , p =2) * torch.norm(t2, p = 2))
  print("cosine_similarity: {}".format(cosine_similarity))

  # TODO 5-8) ì½”ì‚¬ì¸ ìœ ì‚¬ë„ë¥¼ ë°˜í™˜í•´ ì£¼ì„¸ìš”.
  return cosine_similarity # TODO 5-8
````

````python
t1 = torch.tensor([1., 2., 0., 3., 5.])
t2 = torch.tensor([2., 1., 1., 1., 0.])

# TODO 5-9) ìœ„ì—ì„œ ì •ì˜í•œ ë§¨í•´íŠ¼ ìœ ì‚¬ë„ë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜ë¡œ t1, t2 ë‘ Tensorì˜ ë§¨í•´íŠ¼ ìœ ì‚¬ë„ë¥¼ êµ¬í•´ ì£¼ì„¸ìš”. | ë³€ìˆ˜ëª…: manhattan_s
# TODO 5-9
manhattan_s = get_manhattan_similarity(t1, t2)

# TODO 5-10) ìœ„ì—ì„œ ì •ì˜í•œ ìœ í´ë¦¬ë“œ ìœ ì‚¬ë„ë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜ë¡œ t1, t2 ë‘ Tensorì˜ ìœ í´ë¦¬ë“œ ìœ ì‚¬ë„ë¥¼ êµ¬í•´ ì£¼ì„¸ìš”. | ë³€ìˆ˜ëª…: euclidean_s
# TODO 5-10
euclidean_s = get_euclidean_similarity(t1, t2)

# TODO 5-11) ìœ„ì—ì„œ ì •ì˜í•œ ì½”ì‚¬ì¸ ìœ ì‚¬ë„ë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ t1, t2 ë‘ Tensorì˜ ì½”ì‚¬ì¸ ìœ ì‚¬ë„ë¥¼ êµ¬í•´ ì£¼ì„¸ìš”. | ë³€ìˆ˜ëª… cosine_s
# TODO 5-11
cosine_s = get_cosine_similarity(t1, t2)

if manhattan_s <= 0.1 and euclidean_s <= 0.2 and cosine_s <= 0.5:
  print("\nğŸ‰ğŸ‰ğŸ‰ ì„±ê³µ!!! ğŸ‰ğŸ‰ğŸ‰")
else:
  print("\nğŸ˜° ë‹¤ì‹œ ë„ì „í•´ë´ìš”!")
````

````python
def add_random_noise(img_t, scale=30):
  # TODO 7-1) img_tì™€ ëª¨ì–‘(shape)ì´ ê°™ì€ í‘œì¤€ì •ê·œë¶„í¬ ë‚œìˆ˜ Tensorë¥¼ ìƒì„±í•´ ì£¼ì„¸ìš”. | ë³€ìˆ˜ëª… random_noise
  # TODO 7-1
  img_t = img_t.to(torch.float32)
  random_noise = torch.randn_like(img_t)

  print("random_noise.mean: {:.4f}".format(random_noise.mean()))
  print("random_noise.std: {:.4f}".format(random_noise.std()))
  print("random_noise.min: {:.4f}".format(random_noise.min()))
  print("random_noise.max: {:.4f}\n".format(random_noise.max()))

  random_noise *= scale

  # TODO 7-2) img_tì˜ ìë£Œí˜•ì„ 32-bit floating pointë¡œ ë³€í™˜í•œ í›„ random_noiseë¥¼ ë”í•´ ì£¼ì„¸ìš”. | ë³€ìˆ˜ëª…: noise_added_img
  # TODO 7-2
  noise_added_img = img_t.type(torch.float32) + random_noise

  # TODO 7-3) noise_added_imgì˜ ëª¨ë“  ê°’ì´ 0ê³¼ 255 ì‚¬ì´ì˜ ê°’ì„ ê°€ì§€ë„ë¡ ì œí•œí•´ ì£¼ì„¸ìš”. hint: clip
  # TODO 7-3
  noise_added_img = torch.clip(noise_added_img, min = 0, max = 255)

  # TODO 7-4) noise_added_imgì˜ ìë£Œí˜•ì„ torch.uint8ë¡œ ë³€í™˜í•´ ì£¼ì„¸ìš”. hint: Tensor.type()
  # TODO 7-4
  noise_added_img = noise_added_img.type(torch.uint8)

  print("noise_added_img.min: {}".format(noise_added_img.min()))
  print("noise_added_img.max: {}\n".format(noise_added_img.max()))

  if noise_added_img.max() <= 255 and noise_added_img.min() >= 0:
    print("ğŸ‰ğŸ‰ğŸ‰ ì„±ê³µ!!! ğŸ‰ğŸ‰ğŸ‰")
  else:
    print("ğŸ™… ë‹¤ì‹œ ë„ì „í•´ë´ìš”!")
  return noise_added_img.numpy()
````
